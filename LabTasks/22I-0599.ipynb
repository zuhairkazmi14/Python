### Kind

* 'bar': Vertical bar plot.
* 'barh': Horizontal bar plot.
* 'hist': Histogram.
* 'box': Box plot.
* 'kde': Kernel Density Estimation plot.
* 'density': Same as 'kde'.
* 'area': Area plot.
* 'scatter': Scatter plot.
* 'hexbin': Hexagonal bin plot.

### Marker
* 'o': Circle marker (default).
* 's': Square marker.
* '^': Triangle marker.
* '*': Star marker.
* '+': Plus marker.
* 'x': X marker.
* 'D': Diamond marker.
* '.': Point marker.

```python
img_scaled = img * 255

# Display the scaled image
plt.imshow(img_scaled.astype('uint8'))

# Add a colorbar for reference (optional)
plt.colorbar()

# Add title
plt.title('Image Example (Scaled to [0, 255])')

# Show the image
plt.show()
```

## Histogram

```python
# Convert the image to grayscale (assuming it's a color image)
img_gray = np.mean(img_scaled, axis=2)

# Flatten the 2D image array to a 1D array
img_flat = img_gray.flatten()

# Create a histogram
plt.hist(img_flat, bins=256, range=(0, 255), color='b')

# Add labels and title
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.title('Histogram of the Image')

# Show the histogram
plt.show()
```

### Histogram Equilization

```python
# Load an image
image = cv2.imread('w3.png', cv2.IMREAD_GRAYSCALE)

# Plot the original and equalized images side by side
plt.figure(figsize=(10, 5))
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.show()
```

```python
# Load an image
image = cv2.imread('w3.png', cv2.IMREAD_GRAYSCALE)

# Apply histogram equalization
equalized_image = cv2.equalizeHist(image)

# Plot the original and equalized images side by side
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(122)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')

plt.show()
```

```python
# Create a figure with subplots for the images and histograms
fig, axes = plt.subplots(1, 2, figsize=(15, 5))


# Plot the histogram of the original image
axes[0].hist(image.ravel(), bins=256, range=(0, 256), color='b')
axes[0].set_xlabel('Pixel Value')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Histogram of Original Image')

# Plot the histogram of the equalized image
axes[1].hist(equalized_image.ravel(), bins=256, range=(0, 256), color='b')
axes[1].set_xlabel('Pixel Value')
axes[1].set_ylabel('Frequency')
axes[1].set_title('Histogram of Equalized Image')

plt.tight_layout()
plt.show()
```

* Read image again in greyscale this time. Take middle two rows of image and draw a line chart between them. Use at least two different kind and marker

```python
# YOUR CODE HERE
image = cv2.imread('cat.jpg', cv2.IMREAD_GRAYSCALE)

# Get the middle two rows of the image
middle_rows = image[image.shape[0] // 2 - 1:image.shape[0] // 2 + 1, :]

# Create x-axis values (columns)
x = np.arange(0, middle_rows.shape[1])

# Create a figure and axes for the line chart
fig, ax = plt.subplots()

# Plot the first line with a solid line style and 'o' markers
ax.plot(x, middle_rows[0], label='Row 1', linestyle='-', marker='o')

# Plot the second line with a dashed line style and 'x' markers
ax.plot(x, middle_rows[1], label='Row 2', linestyle='--', marker='x')

# Set labels and title
ax.set_xlabel('Column')
ax.set_ylabel('Pixel Value')
ax.set_title('Pixel Value vs. Column for Middle Two Rows')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```

* display only one channel of your image and it's histogram as well

```python
image_path = 'cat.jpg'  # Replace with the path to your image
image = cv2.imread(image_path)

# Extract the Red channel
red_channel = image[:, :, 0]

# Display the Red channel image
plt.subplot(121)
plt.imshow(red_channel, cmap='gray')
plt.title('Red Channel')
plt.axis('off')

# Compute the histogram of the Red channel
histogram = cv2.calcHist([red_channel], [0], None, [256], [0, 256])

# Plot the histogram
plt.subplot(122)
plt.plot(histogram, color='red')
plt.title('Red Channel Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

* Take first column of one channel of your image and second column of third chanel of your image,

* Draw their heatmap

```python

# Load the image using OpenCV
image_path = 'cat.jpg'  # Replace with the path to your image
image = cv2.imread(image_path)

# Extract the Red channel and Blue channel
red_channel = image[:, :, 0]
blue_channel = image[:, :, 2]

# Take the first column of the Red channel and the second column of the Blue channel
column1_red = red_channel[:, 0]
column2_blue = blue_channel[:, 1]

# Create a heatmap by combining the two columns
heatmap = np.column_stack((column1_red, column2_blue))

# Display the heatmap
plt.imshow(heatmap, cmap='hot', aspect='auto')
plt.title('Heatmap of First Column of Red Channel and Second Column of Blue Channel')
plt.colorbar()

# Show the plot
plt.show()
```
